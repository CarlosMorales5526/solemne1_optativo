## Generando la limpieza

En este datasets se generara la siguiente limpieza. Se eliminara las columnas sin importancia ya mencionadas. Se buscara los valores nulos existentes y se cambiaran por el promedio de la columna en la que se encuentra el valor nulo. Y se eliminaran outliers revisando mapas de dispersion

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_frame = pd.read_csv('Placement_Data_Full_Class.csv', sep=',', )
data_frame
```

```python
#dimensiones del datasets
data_frame.shape
```

```python
#tipos de datos
data_frame.dtypes
```

```python
#Revisar nulos
for dato in data_frame.columns:
    print('Total de valores nulos de', dato, '=', data_frame[dato].isna().sum())
```

Eliminacion de columna que no sirve para el caso, la columna del numero serial, la columna de la puntacion de la educacion secundaria y educacion superior. Y las columnas del tipo de junta del directivo de la educacion secundaria y superior

```python
data_frame = data_frame.drop('sl_no', axis = 'columns')
data_frame = data_frame.drop('ssc_p', axis = 'columns')
data_frame = data_frame.drop('ssc_b', axis = 'columns')
data_frame = data_frame.drop('hsc_p', axis = 'columns')
data_frame = data_frame.drop('hsc_b', axis = 'columns')
data_frame = data_frame.drop('hsc_s', axis = 'columns')

data_frame.head()
```

Transformacion de valores de objetos a valores enteros

```python
# transformacion de la columna gender M:0 F:1

d_gender = { 'M': 0, 'F': 1}
data_frame['gender'] = data_frame['gender'].map(d_gender)


#transformacion de la columna de tipo de titulo (Sci&Tech = 0, Comm&Magmt = 1)
d_tipo_titulo = { 'Sci&Tech': 0, 'Comm&Mgmt': 1}
data_frame['degree_t'] = data_frame['degree_t'].map(d_tipo_titulo)

#transformacion de la columna de experiencia laboral (No = 0, Yes = 1)
d_exp_labo = { 'No': 0, 'Yes': 1}
data_frame['workex'] = data_frame['workex'].map(d_exp_labo)

#transformacion de la columna de postgrado o especializado (Mkt&HR = 0, Mkt&Fin = 1)
d_especializado = { 'Mkt&HR': 0, 'Mkt&Fin': 1}
data_frame['specialisation'] = data_frame['specialisation'].map(d_especializado)

#transformacion de la columna estado colocado o no (Placed = 0, Not Placed = 1)
d_colocado = { 'Not Placed': 0, 'Placed': 1}
data_frame['status'] = data_frame['status'].map(d_colocado)

data_frame
```

```python
data_frame.describe()
```

```python
#Se agrega una columna con los datos del salario sin nulos. Se cambian los nulos por el promedio de la columna
data_frame['salary_noNaN'] = data_frame['salary'].fillna(data_frame['salary'].mean())
data_frame
```

# Sección 3 - Regresión


A continuación, aplica un modelo de Regresión o de Clasificación al dataset de los Salarios en Python.

---



```python
import seaborn as sns
from sklearn import linear_model , model_selection
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
```

```python
## Correlación de las variables
plt.figure(figsize=(15,15))
p=sns.heatmap(data_frame.corr(method="pearson"), annot=True,cmap='RdYlGn',square=True)
```

Se realizara una regresion lineal, para poder predecir los valores del salario segun la columna puntaje obtenido en la prueba de especialidad. Por que esta columnas presentan una mayor relacion, segun el mapa de calor, con el salario obtenido.

```python
print(data_frame.groupby('etest_p').size())
```

```python
X = data_frame.etest_p
y = data_frame.salary_noNaN
print("Shape of X :", X.shape, "Shape of y:", y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=29)
X_train = np.array(X_train).reshape(-1,1)
y_train = np.array(y_train).reshape(-1,1)
X_test = np.array(X_test).reshape(-1,1)
y_test = np.array(y_test).reshape(-1,1)

X_train.shape
```

```python
# creando el modelo lineal
modelo = linear_model.LinearRegression()
```

```python
# entrenando al modelo
modelo.fit(X_train, y_train)
```

```python
# generando las predicciones
y_pred = modelo.predict(X_test)
```

```python
# Evaluando el modelo
print('Coeficientes: \n\tb1 =', modelo.coef_, 'a =', modelo.intercept_)
# la media del cuadrado del error
print('Mean squared error (MSE): %.2f' % mean_squared_error(y_test, y_pred))
# coeficiente de determinacion
print('Coeficiente de determinación: %.2f' % r2_score(y_test, y_pred))
print('Score del modelo (train): %.2f' % modelo.score(X_train, y_train)) # Return the coefficient of determination  of the prediction.
print('Score del modelo (test): %.2f' % modelo.score(X_test, y_test)) 
```
